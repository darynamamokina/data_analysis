#### **Система Лоренца:**

Система Лоренца описується трьома диференціальними рівняннями:

$$
\begin{cases}
\frac{dx}{dt} = \sigma (y - x) \\
\frac{dy}{dt} = x (\rho - z) - y \\
\frac{dz}{dt} = xy - \beta z
\end{cases}
$$

де:
- $( \sigma )$ — коефіцієнт Прандтля (зазвичай $ \sigma = 10 $),
- $( \rho )$ — числовий параметр для турбулентності (зазвичай $ \rho = 28 $),
- $( \beta )$ — коефіцієнт відношення висоти до довжини (зазвичай $ \beta = 8/3 $).

#### **Об'єкти Нарендра-Пархтізаратхі:**

Об'єкти Нарендра-Пархтізаратхі описуються наступними рівняннями:

$$
\begin{cases}
x_1(t+1) = \alpha_1 x_1(t) + \beta_1 x_2(t) + \gamma_1 x_3(t) + \delta_1 \\
x_2(t+1) = \alpha_2 x_1(t) + \beta_2 x_2(t) + \gamma_2 x_3(t) + \delta_2 \\
x_3(t+1) = \alpha_3 x_1(t) + \beta_3 x_2(t) + \gamma_3 x_3(t) + \delta_3
\end{cases}
$$

де $(\alpha_i)$, $(\beta_i)$, $(\gamma_i)$ і $(\delta_i)$ — коефіцієнти для кожного з рівнянь.

Їх поведінка моделюється нелінійними диференціальними рівняннями, що описують динаміку в часі. Ці рівняння створюють цікаву, нелінійну та хаотичну поведінку системи.

### **Нелінійних рівнянь Нарендра-Партхізаратхі:**

8. $$
\dot{x}_1 = a_1 \sin(x_2) - b_1 x_1^2
$$
$$
\dot{x}_2 = -a_2 x_2 + b_2 x_1 \cos(x_2)
$$


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import sklearn
import seaborn as sns
from scipy.integrate import odeint
```

```python
def lorenz_system(state, t, sigma=10, rho=28, beta=8/3):
    x, y, z = state
    dx_dt = sigma * (y - x)
    dy_dt = x * (rho - z) - y
    dz_dt = x * y - beta * z
    return [dx_dt, dy_dt, dz_dt]

def narendra_parthasarathy_system(state, t, a1, b1, a2, b2):
    x1, x2 = state
    dx1_dt = a1 * np.sin(x2) - b1 * x1**2
    dx2_dt = -a2 * x2 + b2 * x1 * np.cos(x2)
    return [dx1_dt, dx2_dt]


initial_state_lorenz = [1.0, 1.0, 1.0]
initial_state_np = [1.0, 1.0]
time_points = np.linspace(0, 50, 10000)

lorenz_data = odeint(lorenz_system, initial_state_lorenz, time_points)

a1, b1, a2, b2 = 1.0, 0.5, 2.0, 1.0
np_data = odeint(narendra_parthasarathy_system, initial_state_np, time_points, args=(a1, b1, a2, b2))

fig = plt.figure(figsize=(12, 5))

ax1 = fig.add_subplot(121, projection='3d')
ax1.plot(lorenz_data[:, 0], lorenz_data[:, 1], lorenz_data[:, 2], label='Lorenz System', color='b')
ax1.set_title('Lorenz System')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('Z')
ax1.legend()

ax2 = fig.add_subplot(122)
ax2.plot(np_data[:, 0], np_data[:, 1], label='Narendra-Parthasarathy System', color='r')
ax2.set_title('Narendra-Parthasarathy System (2D)')
ax2.set_xlabel('X1')
ax2.set_ylabel('X2')
ax2.legend()
ax2.grid(True)

plt.tight_layout()
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
np_data = np.hstack((np_data, np.zeros((np_data.shape[0], 1))))
X = np.concatenate((lorenz_data, np_data))
Y = np.concatenate((np.zeros(lorenz_data.shape[0]), np.ones(np_data.shape[0])))  

X_train_val, X_test, Y_train_val, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
X_train, X_val, Y_train, Y_val = train_test_split(X_train_val, Y_train_val, test_size=0.2, random_state=42)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)
X_test_scaled = scaler.transform(X_test)

print(f"Training set size: {X_train_scaled.shape}")
print(f"Validation set size: {X_val_scaled.shape}")
print(f"Test set size: {X_test_scaled.shape}")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import classification_report
```

```python
logistic_reg = LogisticRegression(random_state=42)
decision_tree = DecisionTreeClassifier(random_state=42)
random_forest = RandomForestClassifier(random_state=42)
svm_model = SVC(random_state=42)

logistic_reg.fit(X_train_scaled, Y_train)
decision_tree.fit(X_train_scaled, Y_train)
random_forest.fit(X_train_scaled, Y_train)
svm_model.fit(X_train_scaled, Y_train)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeRegressor
```

```python
param_grid_dt = {
    'max_depth': [3, 5, 10],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
decision_tree = DecisionTreeRegressor()
grid_search_dt = GridSearchCV(decision_tree, param_grid_dt, cv=5, scoring='neg_mean_squared_error', verbose=1)
grid_search_dt.fit(X_train_scaled, Y_train)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
param_grid_rf = {
    'n_estimators': [50, 100, 200],
    'max_depth': [3, 5, 10],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

random_forest = RandomForestRegressor()
grid_search_rf = GridSearchCV(random_forest, param_grid_rf, cv=5, scoring='neg_mean_squared_error', verbose=1)
grid_search_rf.fit(X_train_scaled, Y_train)
```

```python
from sklearn.svm import SVR
```

```python

param_grid_svm = {
    'C': [0.1, 1, 10],
    'kernel': ['linear', 'rbf', 'poly'],
    'gamma': ['scale', 'auto']
}

svm_model = SVR()
grid_search_svm = GridSearchCV(svm_model, param_grid_svm, cv=5, scoring='neg_mean_squared_error', verbose=1)
grid_search_svm.fit(X_train_scaled, Y_train)

```

```python
print("\nBest parameters for Decision Tree:", grid_search_dt.best_params_)
print("Best parameters for Random Forest:", grid_search_rf.best_params_)
print("Best parameters for SVM:", grid_search_svm.best_params_)

optimal_params_tree = grid_search_dt.best_params_  
optimal_params_forest = grid_search_rf.best_params_ 
optimal_params_svr = grid_search_svm.best_params_
```

```python
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
import pandas as pd
import numpy as np

def display_metrics(df_name, models, y):
    result = {}
    
    for key, predict in models.items():
        result[key] = {
            "RMSE": np.sqrt(mean_squared_error(y, predict)),
            "R2": r2_score(y, predict),
        }

    metrics_dataset = pd.DataFrame(result).T 

    print(df_name)
    display(metrics_dataset)

```

```python

validation_prediction = {}

linear_model = LinearRegression().fit(X_train, Y_train)
linear_predict = linear_model.predict(X_val)

final_tree_model = DecisionTreeRegressor(**optimal_params_tree).fit(X_train, Y_train)
tree_predict = final_tree_model.predict(X_val)

final_forest_model = RandomForestRegressor(**optimal_params_forest).fit(X_train, Y_train)
forest_predict = final_forest_model.predict(X_val)

final_svr_model = SVR(**optimal_params_svr).fit(X_train, Y_train)
svr_predict = final_svr_model.predict(X_val)

validation_prediction["Linear Regression"] = linear_predict
validation_prediction["Random Forest"] = forest_predict
validation_prediction["Decision Tree"] = tree_predict
validation_prediction["SVR"] = svr_predict

display_metrics("Validation:", validation_prediction, Y_val)

```

```python
test_prediction = {}

linear_predict = linear_model.predict(X_test)
tree_predict = final_tree_model.predict(X_test)
forest_predict = final_forest_model.predict(X_test)
svr_predict = final_svr_model.predict(X_test)

test_prediction["Linear Regression"] = linear_predict
test_prediction["Random Forest"] = forest_predict
test_prediction["Decision Tree"] = tree_predict
test_prediction["SVR"] = svr_predict

display_metrics("Test:", test_prediction, Y_test)
```
